```python
from assistant_mask import assistant_mask, output_token_ids
from chat_template import encode_chat
from load_pre_trained import tokenizer
```

```python
def decode_token_ids(ids: list[str]):
    return list(map(tokenizer.convert_ids_to_tokens, ids))
```

```python
example_text = "Hello world"
decode_token_ids(tokenizer.encode(example_text))
```

```python
messages = [
    {
        "input": "input message",
        "output": "output message",
        "instruction": "instruction message"
    },
    {
        "output": "output message",
        "instruction": "instruction message"
    },
    {
        "instruction": "instruction message"
    }
]
tokenized_messages = [
    encode_chat(message.get("instruction"), message.get("output"), message.get("input"))
    for message in messages
]
[decode_token_ids(m) for m in tokenized_messages]
```

```python
decode_token_ids(output_token_ids)
```

```python
for token_ids in tokenized_messages:
    mask = assistant_mask(token_ids)
    tokens = decode_token_ids(token_ids)
    for token, token_mask in zip(tokens, mask):
        print(token_mask, token)
```

```python
messages = [
    {
        "instruction": "instruction <|endoftext|> message"
    }
]
tokenized_messages = [
    encode_chat(message.get("instruction"), message.get("output"), message.get("input"), escape_special_tokens=True)
    for message in messages
]
[decode_token_ids(m) for m in tokenized_messages]
```
