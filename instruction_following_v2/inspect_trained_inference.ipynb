```python
import torch

from generation import ChatCompletion
from load_pre_trained import model, tokenizer

device = "cuda" if torch.cuda.is_available() else "cpu"
print(device)
```

```python
model.load_state_dict(torch.load("output/state/014949.pt").get("model"))
model.to(device)
```

```python
def inference(x: torch.Tensor):
    return model(x).logits
```

```python
chat = ChatCompletion(
    tokenizer=tokenizer,
    inference=inference,
    device=device,
    stop_token_ids=[model.config.eos_token_id],
    max_context_length=model.config.n_positions,
    top_p=0.9,
    top_k=None,
    temperature=0.7,
    max_new_tokens=200
)
```

```python
chat.generate(instruction="Hello")
```

```python
chat.generate(instruction="What is the capital of Germany?")
```

```python
chat.generate(instruction="Write a poem about cats.")
```
