```python
import torch
from datasets import load_from_disk

from load_pre_trained import model, tokenizer
from training_data_reader import TrainingDataReader
```

```python
ds = load_from_disk("tokenized_data/test")
```

```python
eos_id = model.config.eos_token_id
device = "cuda" if torch.cuda.is_available() else "cpu"
```

```python
reader = TrainingDataReader(
    context_length=1024,
    padding_token_id=eos_id,
    batch_size=2,
    test_dataset=ds,
    train_dataset=ds,
    device=device,
    ignore_index=-100
)
```

```python
x, y = reader.get_batch()
```

```python
print(x.shape, y.shape)
```

```python
train = x[0]
test = y[0]
mask_id = tokenizer.convert_tokens_to_ids("*")
tokenizer.decode(train)
```

```python
tokenizer.decode([t if t != -100 else mask_id for t in test])
```
