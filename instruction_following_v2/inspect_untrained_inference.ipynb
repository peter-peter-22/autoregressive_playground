# Chat

```python
chat=ChatCompletion(
    tokenizer=tokenizer,
    model=model,
    device=device,
    stop_token_ids=[model.config.eos_token_id],
    max_context_length=model.config.n_positions,
    top_p=0.9,
    top_k=None,
    temperature=0.7,
    max_new_tokens=200
)
```

```python
chat.generate(instruction="Briefly describe gravity.")
```
