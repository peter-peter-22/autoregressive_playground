```python
import torch
import torch.nn as nn

from checkpoints import Checkpointer, CheckpointCleaner
```

```python
class Model(nn.Module):
    def __init__(self):
        super(Model, self).__init__()
        self.layer=nn.Linear(1,1)

    def forward(self,x:torch.Tensor):
        return self.layer(x)

model=Model()
```

```python
checkpoint_dir="temp"
total_steps = 10
eval_interval = 2
log_intervals=1
checkpointer=Checkpointer(
    model=model,
    optimizer=torch.optim.Adam(model.parameters(), lr=1e-3),
    scaler=None,
    checkpoint_dir=checkpoint_dir,
    cleaner=CheckpointCleaner(3),
    max_context_length=100,
    batch_size=2,
    eval_interval=eval_interval
)
```

```python
checkpointer.preserve_progress([0.5],eval_interval,total_steps)
```

```python
for step in range(total_steps):
    if step % log_intervals == 0:
        checkpointer.create_log(step)
    if step % eval_interval == 0 or step == total_steps - 1:
        checkpointer.save(step,step,step,[],0,"")
```
