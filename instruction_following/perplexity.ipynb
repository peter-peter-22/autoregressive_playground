A simple experiment with the calculation of perplexity.

```python
import torch
import torch.nn as nn
```

```python
vocab_size = 10
```

```python
loss = nn.functional.cross_entropy(
    torch.tensor([10, *[-10 for _ in range(vocab_size - 1)]], dtype=torch.float),
    torch.tensor([0], dtype=torch.long)
)
print("loss", loss.item())
print("perplexity", torch.exp(loss).item())
```

```python
loss = nn.functional.cross_entropy(
    torch.tensor([1, *[1 for _ in range(vocab_size - 1)]], dtype=torch.float),
    torch.tensor([0], dtype=torch.long)
)
print("loss", loss.item())
print("perplexity", torch.exp(loss).item())
```
