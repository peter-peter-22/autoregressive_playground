```python
from model import ChatModel

model = ChatModel(
    vocabulary_size=2000,
    embedding_size=64,
    max_context_length=200,
    ff_size_multiplier=2,
    transformer_blocks=4,
    attention_heads=4,
    dropout=0.0,
    bias=True,
    device="cpu",
)
```

```python
for name, param in model.named_parameters():
    print(name)
```

```python
def freeze_lower_layers(num_freeze):
    # embeddings
    model.emb.requires_grad_(False)

    # transformer blocks
    for i in range(num_freeze):
        model.transformer[i].requires_grad_(False)


freeze_lower_layers(2)
```

```python
trainable_params = [n for n, p in model.named_parameters() if p.requires_grad]
print(len(trainable_params))
print(trainable_params)
```

```python
un_trainable_params = [n for n, p in model.named_parameters() if not p.requires_grad]
print(len(un_trainable_params))
print(un_trainable_params)
```
