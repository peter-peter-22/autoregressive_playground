```python
from tokenizers import Tokenizer
from datasets.formatting.formatting import LazyBatch
from datasets import load_dataset
from special_tokens import special_tokens

file_name = "./tokenizer.json"
tokenizer = Tokenizer.from_file(file_name)
```

```python
def tokenize_wiki(batch: LazyBatch):
    eot = special_tokens["end_of_text"]
    results = [
        tokenizer.encode(row + "\n" + eot).ids
        for row in batch["summary"]
    ]
    return {"tokens": results}
```

```python
ds = load_dataset("jordiclive/wikipedia-summary-dataset", split="train").select_columns(["summary"]).take(1)
data=next(iter(ds))
data
```

```python
encoding = tokenizer.encode(data["summary"])
print(encoding.tokens)
decoding = tokenizer.decode(encoding.ids)
print(decoding)
```
