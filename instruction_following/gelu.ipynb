```python
import matplotlib.pyplot as plt
import torch
from torch.nn import GELU

# Generate x values
x = [x / 10 for x in range(-50, 50)]

# Define a function for y (e.g., sine wave)
y = GELU().forward(torch.Tensor(x)).tolist()

# Plot the curve
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('GELU Curve')
plt.grid(True)
plt.show()
```
