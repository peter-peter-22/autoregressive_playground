```python
import torch
import torch.nn as nn
```

```python
logits = torch.tensor([1, 0.9, 0.8, 0.2, 0.1, 0.1])
probs = torch.softmax(logits, dim=0)
torch.argmax(probs, dim=-1)
```

Multinominal

```python
ids = [torch.multinomial(probs, num_samples=1).item() for i in range(1000)]
counts = torch.bincount(torch.tensor(ids))
counts
```

Temperature

```python
def temperature(logits, temp):
    probs = nn.functional.softmax(logits / temp, dim=0)
    ids = [torch.multinomial(probs, num_samples=1).item() for i in range(1000)]
    counts = torch.bincount(torch.tensor(ids))
    return counts
```

```python
temperature(logits, 1)
```

```python
temperature(logits, 0.1)
```

```python
temperature(logits, 10)
```

Top k

```python
top_logits, top_pos=torch.topk(logits,k=3)
top_logits
```

```python
def top_k(logits,top_k):
    top_logits, top_pos=torch.topk(logits,top_k)
    filtered=torch.where(
        logits<top_logits[-1],
        input=torch.tensor(float("-inf")),
        other=logits
    )
    return filtered
```

```python
top_k(logits,3)
```
