```python
from datasets import Dataset, load_from_disk
from millify import millify
from tabulate import tabulate

dataset_list: list[Dataset] = [
    load_from_disk("tokenized_data/train_chats"),
    load_from_disk("tokenized_data/test_chats"),
]
```

```python
data = []
header = ["name", "rows", "bytes", "rows %", "bytes %"]
for dataset in dataset_list:
    data.append([
        dataset.info.dataset_name or "unknown",
        dataset.num_rows,
        dataset.size_in_bytes or dataset.data.nbytes
    ])

total_rows = sum([row[1] for row in data])
total_bytes = sum([row[2] for row in data])
data = [[name, millify(rows), millify(size), round(rows * 100 / total_rows), round(size * 100 / total_bytes)] for
        name, rows, size in data]

print(tabulate(data, header, tablefmt="grid"))
print(f"\n\nTotal bytes: {millify(total_bytes)}, Total rows: {millify(total_rows)}")
```
