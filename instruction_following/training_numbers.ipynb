```python
from math import sqrt

from millify import millify
```

# Chinchilla rule
Optimal training tokens = 20 * model parameters

```python
millify(40_000_000 * 20)
```

# Learning rate
Peak LR = 0.001 * sqrt(batch_size/256)

```python
batch_size = 32
lr = 0.001 * sqrt(batch_size / 256)
print(f"{lr:.2}")
```

# Learning rate schedule
- Warm up: 1-3% of total steps
- Cosine decay: to the 10% of the peak

```python
steps = 100_000
warmup = steps * 0.02
min_lr = 0.1 * lr
print(millify(warmup))
print(f"{min_lr:.2}")
```

# Batch size
Larger = more stable learning

# Instruction tuning
- Learning rate: ~0.2 * peak, 5e-5 - 1e-4, no warmup, no cosine decay, early stopping
- Risk of overfitting
- Dropout must be on, otherwise it won't learn anything

# Compute cost
- FLOPs = 6 * parameters * tokens
- RTX A6000 = 40 TFLOPS

```python
flop=6 * 40_000_000 * 3_000_000_000
millify(flop)
```

```python
training_seconds = flop / 40_000_000_000_000
print("seconds:",training_seconds)
print("hours:",training_seconds / 60 / 60) # Real training time: 6h (includes booting and wasted time)
```

# Repetition
- Pre training: max 2-3, 5+ is too much
- Instruction tuning: typically 2-3, 5+ overfits
