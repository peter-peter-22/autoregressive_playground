```python
from string import whitespace

import sentencepiece as spm

spm.SentencePieceTrainer.Train(
    input="input.txt",
    model_prefix="sp",
    vocab_size=8000,
    model_type="bpe",
    character_coverage=1.0,
    remove_extra_whitespaces=False
)
```

```python
import sentencepiece as spm

sp = spm.SentencePieceProcessor()
sp.load("sp.model")

s = "To be, or not to be:\n\nThat is \nthe question."
ids = sp.encode(s)
print(ids)
print(repr(sp.decode(ids)))
for i in ids:
    print(sp.id_to_piece(i))
```
